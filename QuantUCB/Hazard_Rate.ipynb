The hazard rate of an absolutely continuous probability distribution with cdf F is:
$$ h = f/\bar{F},$$
where f is pdf, $\bar{F} = 1 - F$ is survival function associated with F.

This notebook illustrates the hazard rate plot for Gaussian distribution, poisson distribution and beta distribution. 

Some discussion about Hazard rate can be found in [this lecture notes](https://web.stanford.edu/~lutian/coursepdf/unit1.pdf)

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
from scipy.stats import beta
```

## Gaussian Distribution

```python

x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 1000)

# standard 
pdf = norm.pdf(x)
sf = norm.sf(x)
hr1 = pdf/sf
plt.plot(x, hr1, label = 'loc = 0, scale = 1')

# loc = 2, scale = 1
pdf = norm.pdf(x, loc = 2, scale = 1)
sf = norm.sf(x, loc = 2, scale = 1)
hr1 = pdf/sf
plt.plot(x, hr1, label = 'loc = 2, scale = 1')

# loc = -2, scale = 1
pdf = norm.pdf(x, loc = -2, scale = 1)
sf = norm.sf(x, loc = -2, scale = 1)
hr1 = pdf/sf
plt.plot(x, hr1, label = 'loc = -2, scale = 1')

plt.legend()
```

```python
# loc = 2, scale = 10
pdf = norm.pdf(x, loc = 2, scale = 10)
sf = norm.sf(x, loc = 2, scale = 10)
hr1 = pdf/sf
plt.plot(x, hr1, label = 'loc = 2, scale = 10')

# loc = -2, scale = 10
pdf = norm.pdf(x, loc = -2, scale = 10)
sf = norm.sf(x, loc = -2, scale = 10)
hr1 = pdf/sf
plt.plot(x, hr1, label = 'loc = -2, scale = 10')

plt.legend()
```

## Poisson Distribution 

```python
mu = 0.6
x = np.arange(poisson.ppf(0.01, mu), poisson.ppf(0.99, mu))
```

```python
pmf = poisson.pmf(x, mu)
sf = poisson.sf(x, mu)
hr = pmf/sf

plt.plot(x, hr, 'bo', ms=8, label='poisson pmf')
plt.vlines(x, 0, hr, colors='b', lw=5, alpha=0.5)
```

## Beta Distribution

```python
a, b = 2.30984964515, 0.62687954301
x = np.linspace(beta.ppf(0.01, a, b), beta.ppf(0.99, a, b), 1000)
```

```python
pdf = beta.pdf(x,a,b)
sf = beta.sf(x,a,b)
hr = pdf/sf

plt.plot(x, hr)
```
